## Read processed 10K and 10Q data from CSV 

```python
import os 
import pandas as pd
currentDirectory = os.getcwd()
filename = currentDirectory + '/data_10k10q.csv'

base_df = pd.read_csv(filename,sep="|")
```

## Get Close price data

```python
from datetime import datetime
from datetime import timedelta
import yfinance as yf
import matplotlib.pyplot as plt

base_df['close_price'] = 0.0 
base_df['close_price_20'] = 0.0 
for index, row in base_df.iterrows():
    #print(row)
    #print(int(str(row['file_date'])[0:4]))
    print(row['company'] + ': ' + str(row['file_date']))
    year = int(str(row['file_date'])[0:4])
    month = int(str(row['file_date'])[4:6])
    day = int(str(row['file_date'])[6:8])
    start_date = datetime(year, month, day)
    end_date = datetime(year, month, day) + timedelta(days=1)
    
    data = yf.download(row['company'], start = start_date, end = end_date)
    if len(data) != 0:
        base_df['close_price'][index] = data['Close'][0]
    else: 
        base_df['close_price'][index] = 0
    start_date = datetime(year, month, day) + timedelta(days=20)
    end_date = datetime(year, month, day) + timedelta(days=21)
    data = yf.download(row['company'], start = start_date, end = end_date)
    if len(data) != 0:
        base_df['close_price_20'][index] = data['Close'][0]
    else:
        base_df['close_price_20'][index] = 0
    #print(data)
```

```python
base_df
```

```python
df_price = base_df[(base_df.close_price > 0) & (base_df.close_price_20 > 0 )]
```

```python
df_price_count = df_price[['company','doc_type']].groupby(['company'])['doc_type'] \
                             .count() \
                             .reset_index(name='count') \
                             .sort_values(['count'], ascending=False) \
                             .head(5)
```
